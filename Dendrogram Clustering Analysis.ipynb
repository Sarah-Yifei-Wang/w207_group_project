```python
data_avg = data_avg.loc[data_avg!=data_avg.max()]
data_avg.describe()
```

```python
data_np = data_avg.reset_index().to_numpy()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(data_np, 'single')

```

```python
data_avg.index
```

```python
data_np
```

```python
import matplotlib.pyplot as plt

labels = data_avg.index
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(data_np[:,0],data_np[:,1], label='True Position')

for label, x, y in zip(labels, data_np[:, 0], data_np[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
labelList = data_avg.index

plt.figure(figsize=(100, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
cluster_results = fcluster(linked, 2, criterion='maxclust')
cluster_results
```
